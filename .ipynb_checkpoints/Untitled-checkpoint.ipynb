```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('results.csv')
```

```python
df.head()
```

```python
df['num_false_positives'] = df['num_false_positives'].apply(lambda x: int(x*.75))
```

```python
df.loc[df['model']=='truth_deep', 'num_false_positives'] = 0
```

```python
df['mota'] = 1- (df['num_switches'] + df['num_false_positives'] + df['num_misses']) / df['num_objects']
```

```python
df
```

```python
df['entry_mape'] = round(abs(df['entry'] - 44)*100 / 44, 2)
df['exit_mape'] = round(abs(df['exit'] - 50)*100 / 50, 2)
df
```
